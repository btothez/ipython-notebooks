```python
#http://andrew.gibiansky.com/blog/mathematics/cool-linear-algebra-singular-value-decomposition/
%matplotlib inline
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
#from skimage.io import imread
#im = imread("data/isla_bw.png")


#pic = Image.open('data/isla1.jpg').convert('1')
#pic.save('data/isla_bw3.jpg')
#imshow(pic)



"""
# This works
I = np.asarray(Image.open('data/isla_bw.png').convert('L'))
print I.shape
im = Image.fromarray(np.uint8(I))
im.show()
"""

pic = Image.open('data/isla_bw.png').convert('L')
pic.show()
print pic.size
matrix = np.asarray(pic)
print matrix.shape;

U, s, V = np.linalg.svd(matrix, full_matrices=True)
print "Shapes of Pieces"
print (U.shape, V.shape, s.shape)

print "Shorten singular values"
shorty = 10
s = s[:shorty]
S = np.diag(s)
print S.shape
U = U[0:U.shape[0], 0:shorty]
print "U shape"
print U.shape
V = V[0:shorty, 0:V.shape[1]]
print "V shape"
print V.shape
new_matrix = np.dot(U, np.dot(S, V))


im = Image.fromarray(np.uint8(new_matrix))
im.show()
```
